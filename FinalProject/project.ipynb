```python
import pandas as pd
import os
```

```python
cnts = pd.read_csv('/home/imgyuri/BIOINFO1/binfo1-work/read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
# Filter cnt by condition
#We excluded transcripts with low read counts in RNA-seq (<30 raw reads) or low ribosome footprints (<80 raw footprint tags in siLuc library)
filtered_cnts = cnts[(cnts['RNA-siLuc.bam'] >= 30) & (cnts['RNA-control.bam'] >= 30) & (cnts['RPF-siLuc.bam'] >= 80)].copy()
# Calculate Clip_enrichment and rden_change by guided mission 1
filtered_cnts['clip_enrichment'] = filtered_cnts['CLIP-35L33G.bam'] / filtered_cnts['RNA-control.bam']
filtered_cnts['ribosome_density_change'] = (filtered_cnts['RPF-siLin28a.bam'] / filtered_cnts['RNA-siLin28a.bam']) / (filtered_cnts['RPF-siLuc.bam'] / filtered_cnts['RNA-siLuc.bam'])
filtered_cnts.head()
```
